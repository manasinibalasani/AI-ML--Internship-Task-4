```python
import pandas as pd

# Load dataset
df = pd.read_csv("adult.csv")   # update path if needed

# Display first rows
df.head()
```

```python
df.info()
```

```python
numerical_features = [
    'age',
    'fnlwgt',
    'educational-num',
    'capital-gain',
    'capital-loss',
    'hours-per-week'
]
```

```python
categorical_features = [
    'workclass',
    'education',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'gender',
    'native-country'
]
```

```python
target = 'income'
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['education_encoded'] = le.fit_transform(df['education'])
```

```python
df_encoded = pd.get_dummies(
    df,
    columns=categorical_features,
    drop_first=True
)
```

```python
df_encoded[numerical_features].describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_encoded[numerical_features] = scaler.fit_transform(
    df_encoded[numerical_features]
)
```

```python
df_encoded[numerical_features].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.boxplot(data=df_encoded[numerical_features])
plt.xticks(rotation=45)
plt.title("Numerical Features After Scaling")
plt.show()
```

```python
df_encoded.to_csv("adult_preprocessed.csv", index=False)
```
